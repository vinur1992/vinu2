```python
Y = clev1[['num']]
X = clev1.drop(['num'], axis=1)
```

```python
from sklearn.ensemble import ExtraTreesClassifier 
import matplotlib.pyplot as plt 
model = ExtraTreesClassifier() 
model.fit(X,Y)
print(model.feature_importances_) 
#use inbuilt class feature_importances of tree based classifiers 
#plot graph of feature importances for better visualization 
feat_importances = pd.Series(model.feature_importances_, index=X.columns) 
feat_importances.nlargest(10).plot(kind='barh') 
plt.show()
```

```python
clevnew=clev1[['slope','exang','cp','trestbps','chol','thal','age','oldpeak','thalach','ca','num']]
```

```python
Xn=clevnew.drop(['num'], axis=1)
Yn=clevnew[['num']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Xn, Yn, test_size=0.2, random_state=365)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
lr.score(X_test,Y_test)
```

```python
from sklearn.model_selection import KFold 
from sklearn.model_selection import cross_val_score 
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
Xn1 = scaler.fit_transform(Xn)
```

```python
kfold = KFold(n_splits=5, random_state=7) 
model = LogisticRegression() 
scoring = 'accuracy'
results = cross_val_score(model, Xn1, Yn, cv=kfold, scoring=scoring)
print("Accuracy:",results.mean(), results.std()) 
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

model_lr = LogisticRegression()
model_lr.fit(X_train, Y_train)

predictions = model_lr.predict(X_test)
score = model_lr.score(X_test, Y_test)

print("Accuracy = " + str(score))
print(confusion_matrix(Y_test, predictions))
print(classification_report(Y_test, predictions))
```
